```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
def balldata():
  
  data = []
  labels = []
  height, width = 250, 500
  dx, dy = 10, 5
  # x = random.randint(0, width)

  for b in range(10):
    label = random.choice((0, 1))
    frames = []

    if label == 0:

        x = random.randint(0, 10)
        y = height

        for i in range(50):
          img = np.zeros((height, width))
          cv2.circle(img, (x, y), 10, (255), -1)
          x = x + dx
          y = y - dy
          labels.append(label)
          frames.append(img)

    elif label == 1:

        x = random.randint(width-10, width)
        y = 10

        for i in range(50):
          img = np.zeros((height, width))
          cv2.circle(img, (x, y), 10, (255), -1)
          x = x - dx
          y = y + dy
          labels.append(label)
          frames.append(img)

    data.append(frames)

  return data, labels
```

```python
data, labels = balldata()

data = np.array(data)
labels = np.array(labels)
data = data[..., np.newaxis]
labels = labels[..., np.newaxis]

print(data.shape)
print(labels.shape)
```
